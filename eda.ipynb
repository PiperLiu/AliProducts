```python
import matplotlib.pyplot as plt
import pickle
import misc_utils as utils
%matplotlib inline
```

```python
# 训练集是long-tailed问题
with open('train_data.pkl', 'rb') as f:
    data = pickle.load(f, encoding='bytes')

category = list(range(len(data)))
number_of_images = [d[0] for d in data]

first_slice = 500

line, = plt.plot(category[:], number_of_images[:], linewidth=2)
plt.xlabel('category(0-50000)')
plt.ylabel('number of images')
plt.show()
```

```python
# 验证集样本较为均衡，在2-4之间
with open('val_data.pkl', 'rb') as f:
    data = pickle.load(f, encoding='bytes')

category = list(range(len(data)))
number_of_images = [d[0] for d in data]

first_slice = 500


line, = plt.plot(category[:], number_of_images[:], linewidth=2)
plt.xlabel('category(0-50000)')
plt.ylabel('number of images')
plt.show()
```
